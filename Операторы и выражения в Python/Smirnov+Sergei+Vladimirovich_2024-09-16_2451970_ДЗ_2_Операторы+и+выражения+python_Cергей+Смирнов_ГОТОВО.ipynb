Задача 1. Создайте программу, запрашивающую у пользователя два целых числа $a$ и $b$, ввод которых осуществляется через символ пробела, т.е. числа $a$ и $b$ вводятся в одной строке через пробел. Выведите на экран результаты следующих математических операций (предполагается, что соответствующие операции определены):
- сумма `a` и `b`;
- разность `a` и `b`;
- произведение `a` и `b`;
- частное от деления `a` на `b`;
- остаток от деления `a` на `b`;
- десятичный логарифм числа `a;
- результат возведения числа `a` в степень `b`;
- квадратный корень из числа `a`;
- результат умножения числа $e$ на `b`.

Примечание: рекомендуется
- для ввода чисел использовать функцию [`input()`](https://pythonru.com/osnovy/poluchenie-polzovatelskogo-vvoda-v-python-s-input);

Напомним, что подключить библиотеку можно следующим способом
```python

```

````python
import math

# Запрашиваем у пользователя два целых числа через пробел
input_str = input("Введите два целых числа через пробел: ")

# Разделяем строку на два числа
a, b = map(int, input_str.split())

# Выполняем математические операции
sum_result = a + b
difference_result = a - b
product_result = a * b
quotient_result = a / b
remainder_result = a % b
log_result = math.log10(a)
power_result = a ** b
sqrt_result = math.sqrt(a)
e_times_b_result = math.e * b

# Выводим результаты на экран
print(f"Сумма a и b: {sum_result}")
print(f"Разность a и b: {difference_result}")
print(f"Произведение a и b: {product_result}")
print(f"Частное от деления a на b: {quotient_result}")
print(f"Остаток от деления a на b: {remainder_result}")
print(f"Десятичный логарифм числа a: {log_result}")
print(f"Результат возведения числа a в степень b: {power_result}")
print(f"Квадратный корень из числа a: {sqrt_result}")
print(f"Результат умножения числа e на b: {e_times_b_result}")
````

Задача 2. Имеются две пиццы диаметрами 35 см. и 40 см., соответственно. Насколько процентов отличаются площади двух пицц?

Напомним, что площадь круга вычисляется по следующей формуле
$$S = \pi r^2,$$
где $r$ -- радиус круга.

````python
import math

# Диаметры пицц
diameter1 = 35
diameter2 = 40

# Радиусы пицц
radius1 = diameter1 / 2
radius2 = diameter2 / 2

# Вычисляем площади пицц
area1 = math.pi * (radius1 ** 2)
area2 = math.pi * (radius2 ** 2)

# Вычисляем разницу в площадях в процентах
difference_percentage = ((area2 - area1) / area1) * 100

# Выводим результат
print(f"Площадь первой пиццы (диаметр {diameter1} см): {area1:.2f} кв. см.")
print(f"Площадь второй пиццы (диаметр {diameter2} см): {area2:.2f} кв. см.")
print(f"Разница в площадях: {difference_percentage:.2f}%")
````

````python
# Создаем список чисел в интервале (1, 20) с шагом 3
numbers = list(range(1, 20, 3))

# 1. Выводим длину списка
print(f"Длина списка: {len(numbers)}")

# 2. Выводим все элементы списка, начиная со второй позиции
print(f"Элементы, начиная со второй позиции: {numbers[1:]}")

# 3. Выводим все четные элементы списка
even_numbers = [num for num in numbers if num % 2 == 0]
print(f"Четные элементы списка: {even_numbers}")

# 4. Выводим все элементы списка, делящиеся на 5 без остатка
divisible_by_5 = [num for num in numbers if num % 5 == 0]
if divisible_by_5:
    print(f"Элементы, делящиеся на 5: {divisible_by_5}")
else:
    print("Элементов, делящихся на 5, нет")

# 5. Сортируем список в порядке убывания
sorted_descending = sorted(numbers, reverse=True)
print(f"Список в порядке убывания: {sorted_descending}")

# 6. Рассчитываем квадраты элементов исходного списка
squares = list(map(lambda x: x ** 2, numbers))
print(f"Квадраты элементов списка: {squares}")

# 7. Выводим элементы списка в обратном порядке
reversed_numbers = numbers[::-1]
print(f"Элементы списка в обратном порядке: {reversed_numbers}")

# 8. Рассчитываем выборочное среднее
n = len(numbers)
sum_numbers = sum(numbers)
mean = sum_numbers / n
print(f"Выборочное среднее: {mean}")
````

- чему равно модальное значение ([мода](https://ru.wikipedia.org/wiki/%D0%9C%D0%BE%D0%B4%D0%B0_(%D1%81%D1%82%D0%B0%D1%82%D0%B8%D1%81%D1%82%D0%B8%D0%BA%D0%B0)#:~:text=%D0%9C%D0%BE%CC%81%D0%B4%D0%B0%20%E2%80%94%20%D0%B7%D0%BD%D0%B0%D1%87%D0%B5%D0%BD%D0%B8%D0%B5%20%D0%B2%D0%BE%20%D0%BC%D0%BD%D0%BE%D0%B6%D0%B5%D1%81%D1%82%D0%B2%D0%B5%20%D0%BD%D0%B0%D0%B1%D0%BB%D1%8E%D0%B4%D0%B5%D0%BD%D0%B8%D0%B9,%D0%BC%D0%BE%D0%B6%D0%BD%D0%BE%20%D1%81%D0%BA%D0%B0%D0%B7%D0%B0%D1%82%D1%8C%2C%20%D1%87%D1%82%D0%BE%20%D1%81%D0%BE%D0%B2%D0%BE%D0%BA%D1%83%D0%BF%D0%BD%D0%BE%D1%81%D1%82%D1%8C%20%D0%BC%D1%83%D0%BB%D1%8C%D1%82%D0%B8%D0%BC%D0%BE%D0%B4%D0%B0%D0%BB%D1%8C%D0%BD%D0%B0.)) температуры за неделю. В какие дни она наблюдалась?;
- будем считать, что <<день прохладный>>, если температура воздуха строго ниже 14 градусов по шкале Цельсиа, в противном случае <<день теплый>>. Выведите список <<прохладных>> и <<теплых>> дней недели.

````python
# Представление данных в виде словаря
temperatures = {
    "понедельник": 12.0,
    "вторник": 12.0,
    "среда": 17.0,
    "четверг": 17.0,
    "пятница": 14.3,
    "суббота": 11.9,
    "воскресенье": 12.0
}

# Функция для перевода температуры из Цельсия в Фаренгейт
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# 1. Перевод температуры в Фаренгейт и создание кортежей (C, F)
fahrenheit_temperatures = [(day, celsius, celsius_to_fahrenheit(celsius)) for day, celsius in temperatures.items()]
print("Температура по шкале Фаренгейта:")
for day, celsius, fahrenheit in fahrenheit_temperatures:
    print(f"{day}: ({celsius}, {fahrenheit:.2f})")

# 2. Определение дней с максимальной и минимальной температурой
max_temp = max(temperatures.values())
min_temp = min(temperatures.values())
max_days = [day for day, temp in temperatures.items() if temp == max_temp]
min_days = [day for day, temp in temperatures.items() if temp == min_temp]
print(f"Максимальная температура ({max_temp}°C) наблюдалась в днях: {', '.join(max_days)}")
print(f"Минимальная температура ({min_temp}°C) наблюдалась в днях: {', '.join(min_days)}")

# 3. Вычисление среднего значения температуры за неделю
average_temp = sum(temperatures.values()) / len(temperatures)
print(f"Среднее значение температуры за неделю: {average_temp:.2f}°C")

# 4. Определение модального значения температуры и дней, в которые она наблюдалась
from collections import Counter
temp_counts = Counter(temperatures.values())
modal_temp = temp_counts.most_common(1)[0][0]
modal_days = [day for day, temp in temperatures.items() if temp == modal_temp]
print(f"Модальное значение температуры ({modal_temp}°C) наблюдалось в днях: {', '.join(modal_days)}")

# 5. Определение "прохладных" и "теплых" дней
cool_days = [day for day, temp in temperatures.items() if temp < 14]
warm_days = [day for day, temp in temperatures.items() if temp >= 14]
print(f"Прохладные дни: {', '.join(cool_days)}")
print(f"Теплые дни: {', '.join(warm_days)}")
````

Задача 7. Целое положительное число из отрезка $[3;50]$ вводится с клавиатуры. Найдите сумму всех положительных четных чисел, строго меньших введенного.

Примечание: здесь можно предложить несколько решений, например, только через списки или с использованием циклов. В целях ознакомления и овладения навыками работы с циклами рекомендуется использовать цикл с предусловием `while`.

````python
# Запрашиваем у пользователя целое положительное число в диапазоне от 3 до 50
while True:
    try:
        number = int(input("Введите целое положительное число от 3 до 50: "))
        if 3 <= number <= 50:
            break
        else:
            print("Число должно быть в диапазоне от 3 до 50. Попробуйте еще раз.")
    except ValueError:
        print("Введите корректное целое число.")

# Инициализируем переменные для суммы и текущего числа
sum_even = 0
current_number = 2  # Начинаем с первого четного числа

# Используем цикл while для нахождения суммы всех положительных четных чисел, меньших введенного числа
while current_number < number:
    sum_even += current_number
    current_number += 2  # Переходим к следующему четному числу

# Выводим результат
print(f"Сумма всех положительных четных чисел, меньших {number}, равна {sum_even}")
````

````python
# Удаляем знаки препинания и разбиваем текст на слова
import re
words = re.findall(r'\b\w+\b', text_lower)
unique_words = set(words)
num_unique_words = len(unique_words)
print(f"Количество уникальных слов в тексте: {num_unique_words}")
````

````python
from collections import Counter
word_counts = Counter(words)
print("Частотная таблица для каждого уникального слова:")
for word, count in word_counts.items():
    print(f"{word}: {count}")
````

````python
# Извлекаем все буквы из текста
letters = re.findall(r'[а-я]', text_lower)
unique_letters = set(letters)
num_unique_letters = len(unique_letters)
print(f"Количество уникальных букв в тексте: {num_unique_letters}")
````
